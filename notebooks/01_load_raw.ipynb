JSON local → PostgreSQL

```python
# Cellule 1 : Imports
import json
import pandas as pd
from sqlalchemy import create_engine, text

# Cellule 2 : Connexion DB
db_url = "postgresql://game_user:game_password@postgres:5432/game_dw"
engine = create_engine(db_url)

# Cellule 3 : Charger le JSON
with open('/home/jovyan/work/data/raw/match_v5.json', 'r') as f:
    matches_data = json.load(f)

print(f"Nombre de matchs : {len(matches_data)}")

# Cellule 4 : Insérer dans raw_matches
with engine.connect() as conn:
    for match in matches_data:
        conn.execute(
            text("INSERT INTO raw_matches (data) VALUES (:data)"),
            {"data": json.dumps(match)}
        )
    conn.commit()

print("✅ Données brutes chargées dans PostgreSQL")
```
